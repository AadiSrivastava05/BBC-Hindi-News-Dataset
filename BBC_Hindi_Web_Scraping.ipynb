Scraping from the भारत category




```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Function to get the article content
def get_article_content(article_url):
    article_response = requests.get(article_url)
    article_soup = BeautifulSoup(article_response.content, 'html.parser')
    content = ""
    for item in article_soup.find_all('div', dir='ltr'):
        content_div = item.find('p', dir='ltr')
        if content_div:
            content_line = content_div.get_text(separator='\n').strip()
        else:
            content_line = ""
        content += " " + content_line

    if content == "":
        content = "Content not found"
    return content

# Function to scrape headlines and content from a BBC Hindi topic page
def scrape_bbc_hindi(url):
    articles = []
    pg = 1
    url1 = url + f"?page={pg}"

    while url1:
        response = requests.get(url1)
        soup = BeautifulSoup(response.content, 'html.parser')

        page_headlines = []

        # Scrape articles on the current page
        for item in soup.find_all('div', class_='promo-text'):
            headline = item.find('a').get_text().strip()
            article_url = item.find('a')['href']
            if not article_url.startswith('https'):
                article_url = f"https://www.bbc.com{article_url}"
            content = get_article_content(article_url)
            articles.append([headline, content])
            page_headlines.append(headline)

        # Print the first three headlines from the current page
        print(f"Page {pg} complete! First 3 headlines:")
        for i, headline in enumerate(page_headlines[:3]):
            print(f"{i + 1}. {headline}")

        pg += 1
        if pg == 41:  # Adjust the page limit as needed
            break

        url1 = url + f"?page={pg}"
        time.sleep(1)

    return articles

# URL of the BBC Hindi topic page to scrape
url = "https://www.bbc.com/hindi/topics/ckdxnkz7607t"

# Scrape the articles
articles_data = scrape_bbc_hindi(url)

# Save the data to a CSV file
df = pd.DataFrame(articles_data, columns=['Headline', 'Content'])
df.to_csv('bbc_hindi_articles.csv', index=False, encoding='utf-8')

print("Scraping completed and data saved to bbc_hindi_articles.csv")
```
